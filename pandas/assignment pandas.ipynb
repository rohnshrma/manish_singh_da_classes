```python
# ✅ Tasks:
# 	1.	Load CSV using pandas.
# 	2.	Show first & last 5 rows.
# 	3.	Print info() and describe().
# 	4.	Find missing values.
# 	5.	Replace “n/a”, “unknown”, “None” with np.nan.
# 	6.	Drop rows where Email is missing.
# 	7.	Fill missing Age with mean.
# 	8.	Fill missing Score with median.
# 	9.	Fill missing Country with “India”.
# 	10.	Drop rows with more than 3 missing fields.
# 	11.	Clean Name – strip spaces, convert to Title Case.
# 	12.	Standardize Gender (e.g., “M”, “male” → “Male”).
# 	13.	Keep only valid Emails (must contain “@” and “.”).
# 	14.	Clean Phone – remove special chars, keep 10 digits.
# 	15.	Remove duplicate rows.
# 	16.	Remove outliers in Score using IQR.
# 	17.	Convert EnrollmentDate to datetime.
# 	18.	Extract Year & Month into new columns.
# 	19.	Convert Passed to boolean (True/False).
# 	20.	Export cleaned data with selected columns to a new CSV.
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./student_data.csv")
```

```python
print("First Five Rows\n\n")

print(df.head())

print("\n\nlast Five Rows\n\n")
print(df.tail())
```

```python
print("\n****INFO****\n")
print(df.info())

print("\n****DESCRIPTION****\n")
print(df.describe())
```

```python
print("\n****MISSING VALUES****\n")
print(df.isnull().sum())
```

```python
df = df.replace(["n/a" , "unknown" , "None"] , np.nan)
```

```python
df
```

```python
df = df.dropna(subset=["Email"])
```

```python
df
```

```python
print("\n****MISSING VALUES****\n")
print(df.isnull().sum())
```

```python
df.loc[:, "Age"] = df["Age"].round().astype("int64").fillna(df["Age"].mean().round().astype("int64"))
```

```python
df
```

```python
df.loc[: , "Score"] = df["Score"].fillna(df["Score"].median())
```

```python
df
```

```python
df.loc[:,"Country"] = df["Country"].fillna("India")
```

```python
df
```

```python
df = df.dropna(thresh=len(df.columns) - 3)
```

```python
df
```

```python
df.loc[:,"Name"] = df["Name"].str.strip().str.title()
```

```python
df
```

```python
df.loc[:,"Gender"] = df["Gender"].str.lower().replace({"m" : "Male", "male" : "Male","f" : "Female","female" : "Female"})
```

```python
df
```

```python
df = df[df["Email"].str.contains("@.*\.+" ,na=False)]
```

```python
df
```

```python
import re

def clean_num(phone):
    if pd.isna(phone):
        return np.nan
    # digits = re.sub(r"\D", "" , str(phone))
    digits = "".join(filter(str.isdigit , str(phone)))
    if len(digits) >= 10:
        return digits[-10:]
    else:
        np.nan
                     
    
```

```python
df["Phone"] = df["Phone"].apply(clean_num)
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df["EnrollmentDate"] = pd.to_datetime(df["EnrollmentDate"] , errors="coerce")
```

```python
df
```

```python
df["EnrollmentYear"] = df["EnrollmentDate"].dt.year.astype("Int64")
df["EnrollmentMonth"] = df["EnrollmentDate"].dt.month.astype("Int64")
```

```python
df
```

```python
df["Passed"]= df["Passed"].replace({"Yes" : True, "No":False})
```

```python
df
```

```python
df[['Name', 'Age', 'Gender']].to_csv("cleaned_data.csv",index=False)
```
