# Pandas series
- it's like a column in a table
- one dimensional array 

```python
a = [23,45,23]
res = pd.Series(a)
print(res)
```

```python
a = [23,45,23]
res = pd.Series(a , index=["aryan" , "rohan","manish"])
print(res)
```

```python
print(res["aryan"])
```

```python
marks = {
    "john" : 13,
    "kunal" :34,
    "sumit" : 45
}
```

```python
data = pd.Series(marks)
print(data)
```

# DataFrames
- Data Sets in pandas usually multi-dimensional tables, called DataFrames
- DataFrame is a the whole table ,where Series is like a column

```python
data = {
    "names" : ["rohan","john","kenny"],
    "marks" : [34,56,67]
}
```

```python
df = pd.DataFrame(data)

print(df)
```

```python
print(df.loc[0])
```

```python
print(df.loc[[0 , 1]])
```

```python
df = pd.DataFrame(data , index = ["stu1","stu2" , "stu3"])

print(df)
```

```python
print(df.loc[["stu1" , "stu2"]])
```
